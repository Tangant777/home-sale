مدلی که امتیاز حدودا 65 درصد را در کگل به من داد

```python
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv(r'/content/train.csv')
train
```

```python
for i in range(81):
  col=train.iloc[:,i]
  if(col.dtypes=='int64' or col.dtypes=='float64'):
    col.fillna(col.mean(),inplace=True)
  if(col.dtypes=='object'):
    col.fillna(col.mode().iloc[0],inplace=True)
train
```

```python
train.drop(columns='Id',inplace=True)
```

```python
train['SalePrice']
```

```python
from sklearn.preprocessing import LabelEncoder

for i in range(79):
  col=train.iloc[:,i]
  if(col.dtypes=='object'):
    label_encoder = LabelEncoder()
    train.iloc[:,i] = label_encoder.fit_transform(train.iloc[:,i])
train
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
x=train.drop(columns='SalePrice')
y=train['SalePrice']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
reg=LinearRegression()
reg.fit(x_train,y_train)
y_pred=reg.predict(x_test)
plt.scatter(y_test,y_pred)
plt.show()
reg.score(x_test,y_test)
```
